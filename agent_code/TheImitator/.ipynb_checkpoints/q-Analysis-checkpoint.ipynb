```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.signal import savgol_filter
```

```python
def eps_greedy(t,N,C,E):
    return E/(C*t/N+1)
```

```python
N = 100 
C = 10
E = 1
t = np.arange(N)
e = eps_greedy(t,N,C,E)

plt.plot(t, e)
plt.title(r'$\epsilon$-greedy Strategy')
plt.xlabel('Episode')
plt.ylabel(r'$\epsilon$')
plt.show()
```
