```python
r=np.genfromtxt('rewards.txt')
```

```python
window = 10
smoothed_rewards = [np.mean(r[i-window:i+1]) if i > window 
                    else np.mean(r[:i+1]) for i in range(len(r))]

#plt.plot(r)
plt.plot(smoothed_rewards)
```

```python
feature_ind = [1,2,3,5,6,6,6]
action_ideas = ['UP','UP','UP','UP','BOMB', 'DOWN', 'UP', 'LEFT', 'UP', 'DOWN', 'RIGHT']
```

```python
feature_ind = np.array(feature_ind)
```

```python
action_dict = {'RIGHT':0, 'LEFT':1, 'UP':2, 'DOWN':3, 'BOMB':4, 'WAIT':5}
```

```python
features = []
action_ideas = [action_dict[k] for k in action_ideas][4:] #drop first four
for a in range(6):
    f_a = np.zeros(7)
    for i in range(len(feature_ind)):
        if action_ideas[i]==a:
            f_a[feature_ind[i]]=1
    features.append(f_a)
```

```python
print(features)
```
