```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
r=np.genfromtxt('rewards.txt')
```

```python
window = 3
smoothed_rewards = [np.mean(r[i-window:i+1]) if i > window 
                    else np.mean(r[:i+1]) for i in range(len(r))]

#plt.plot(r)
plt.plot(smoothed_rewards)
```
